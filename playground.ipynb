```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
```

```python
#Init a data list:
data_list = []

#Loop over all the files in the data folder:
for file in os.listdir('data/'):
    #Check if the file is a csv file:
    if file.endswith('.csv'):
        #Read the csv file:
        spec_data = pd.read_csv("data/"+file, \
                                delimiter = ',', parse_dates=[0], \
                                    infer_datetime_format=True, na_values='0', \
                                        header = None)
        data_list.append(spec_data)

#Make a dataframe from the data list:
spectral_data = pd.concat(data_list, axis=0, ignore_index=True)
```

```python
spectral_data
```

```python
spectral_data.columns = ['datetime'] + [i for i in range(spectral_data.shape[1] - 1)]
spectral_data.columns
```

```python
spectral_data
```

```python

#Group by 3 hours and take the mean:
spectral_data['datetime'] = pd.to_datetime(spectral_data['datetime'], errors='coerce')  
agg_data = spectral_data.groupby(pd.Grouper(key='datetime', freq='3H')).mean().reset_index()
```

```python
agg_data
```

```python
#Dumping this data to a csv file:
agg_data.to_csv('data/agg_data.csv', index=False) 
```

```python
#Plotting the x, y, and z of the aggeragted magnetic field vector:
sns.lineplot(agg_data['datetime'], agg_data[0], color='red')
sns.lineplot(agg_data['datetime'], agg_data[1], color='blue')
sns.lineplot(agg_data['datetime'], agg_data[2], color='black')

plt.show()
```

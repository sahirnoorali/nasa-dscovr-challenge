```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
```

```python
#Init a data list:
data_list = []

#Loop over all the files in the data folder:
for file in os.listdir('data/'):
    #Check if the file is a csv file:
    if file.endswith('.csv'):
        #Read the csv file:
        spec_data = pd.read_csv("data/"+file, \
                                delimiter = ',', parse_dates=[0], \
                                    infer_datetime_format=True, na_values='0', \
                                        header = None)
        data_list.append(spec_data)

#Make a dataframe from the data list:
spectral_data = pd.concat(data_list, axis=0, ignore_index=True)
```

```python
spectral_data
```

```python
spectral_data.columns = ['datetime'] + [i for i in range(spectral_data.shape[1] - 1)]
spectral_data.columns
```

```python
spectral_data
```

```python

#Group by 3 hours and take the mean:
spectral_data['datetime'] = pd.to_datetime(spectral_data['datetime'], errors='coerce')  
agg_data = spectral_data.groupby(pd.Grouper(key='datetime', freq='3H')).mean().reset_index()
```

```python
agg_data
```

```python
#Dumping this data to a csv file:
agg_data.to_csv('data/agg_data.csv', index=False) 
```

```python
#Plotting the x, y, and z of the aggeragted magnetic field vector:
sns.lineplot(agg_data['datetime'], agg_data[0], color='red')
sns.lineplot(agg_data['datetime'], agg_data[1], color='blue')
sns.lineplot(agg_data['datetime'], agg_data[2], color='black')

plt.show()
```

```python
B_vector = agg_data.loc[:,0:2]#.values
B_vector
```

```python
B_vector = B_vector.dropna()
```

```python
from sklearn.cluster import KMeans

#Fit the kmeans model to the magnetic field vector:
kmeans = KMeans(n_clusters=2, random_state=0, max_iter=1000, n_init=100)
kmeans.fit(B_vector)
```

```python
kmeans.cluster_centers_
```

```python
#Plot the clusters:
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
```

```python
# Import TSNE
from sklearn.manifold import TSNE

# Create a TSNE instance: model
model = TSNE(learning_rate=200)
# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(B_vector)
# Select the 0th feature: xs
xs = tsne_features[:,0]
# Select the 1st feature: ys
ys = tsne_features[:,1]
# Scatter plot, coloring by variety_numbers
plt.scatter(xs,ys, c=kmeans.labels_)
plt.show()
```
